## Collaborative Filtering based recommender system

```python
x = ratings_with_name.groupby('userId').count()['rating'] > 1500
movie_watchers=x[x].index
```

```python
ratings_with_name
```

```python
filtered_rating = ratings_with_name[ratings_with_name['userId'].isin(movie_watchers)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('title').count()['rating']>=150
famous_movies = y[y].index
```

```python
filtered_rating
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_movies)]
```

```python
final_ratings
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_movies)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='title',columns='userId',values='rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
movier = pd.read_csv('movier.csv')
```

```python
movier
```

movie4.head

```python
def recommend(movie_name):
    #index fetch
    index=np.where(pt.index==movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=movier[movier['title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['genres'].values))
        item.extend(list(temp_df.drop_duplicates('title')['tmdbId'].values))
    
    
        data.append(item)
    
    return data
```

```python
print(recommend("(500) Days of Summer "))
```

```python
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(movier,open('movier.pkl','wb'))
pickle.dump(similarity_scores,open('similarity_scores.pkl','wb'))
```
